```python
# Plot a scatter plot between BMI and insurance charges. 
#Write down your observations about the relationship between these two variables.
import pandas as pd
import seaborn as sns
health_data = pd.read_csv("HealthData.csv")
sns.scatterplot(x="bmi", y="charges", data=health_data, hue="smoker")

```

# Observation:

Normal BMI range: 18.5-24.9. From the above plot we see that there is a slight bias/increase in charges for BMI above 30 and it is also evident that smokers incur more charges regardless of BMI. 

For non-smokers, charges remain relatively stable across different BMI values.


```python
#Plot histograms or density plots for the following variables 
#charges , smoker , age: Write down your observations .

import matplotlib.pyplot as plt

fig, axes = plt.subplots(3, 1, figsize=(10, 12))

# Plot histogram for charges
sns.histplot(data=health_data, x='charges', kde=True, ax=axes[0])
axes[0].set_title('Distribution of Insurance Charges')
axes[0].set_xlabel('Charges ($)')
axes[0].set_ylabel('Count')

# Plot count for smoker status
sns.countplot(data=health_data, x='smoker', ax=axes[1])
axes[1].set_title('Distribution of Smoker Status')
axes[1].set_xlabel('Smoker')
axes[1].set_ylabel('Count')

# Plot histogram for age
sns.histplot(data=health_data, x='age', kde=True, ax=axes[2])
axes[2].set_title('Distribution of Age')
axes[2].set_xlabel('Age')
axes[2].set_ylabel('Count')

plt.tight_layout()

```

# Observation

Charges - We have more data on lower charges than the higher ones (right-skewed)

Smoker - The data is highly biased towards the non-smokers

Age - Data is almost equally distributed except that the data for age range 20 is a little higher
